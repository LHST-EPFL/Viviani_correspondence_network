```python
#initialisation of the database
col = ['Sender' , 'Main title', 'Notes', 'in:' , 'Related names', 'subjects', 'places']
bdd =pd.DataFrame(columns= col)

for i in Ref.index.values:
    for i in range(Ref.at[i,'Beginning'] , Ref.at[i,'End']+1, 1):
        url = 'https://opac.museogalileo.it/imss/resource?uri=00000'+"% s" % i
        page = requests.get(url)
        soup = BeautifulSoup(page.content, 'lxml')
        url_ = 'https://opac.museogalileo.it/imss/'
        for link in soup.find_all('iframe'):
            url_ += link['src']

        page = requests.get(url_)
        soup = BeautifulSoup(page.content, 'lxml')
        soup    

        Sender = ''
        Main_title = ''
        Notes = ''
        In = ''
        Related_names = ''
        subjects = ''
        places = ''



        for link in soup.find_all(['a','td']):
            if 'Mittente' in link.next_element:
                Sender = link.next_element.next_element.get_text()
            if 'Titolo' in link.next_element:
                Main_title = link.next_element.next_element.get_text()
            if 'Note:' in link.next_element:
                Notes = link.next_element.next_element.get_text()
            if 'In' in link.next_element:
                In = link.next_element.next_element
            if 'Nomi correlati' in link.next_element:
                Related_names = link.next_element.next_element.get_text()
            if 'Soggetti' in link.next_element:
                subjects = link.next_element.next_element.get_text()
            if 'Luoghi' in link.next_element:
                places = link.next_element.next_element.get_text()
        row = {'Sender' : Sender,'Main title': Main_title,'Notes': Notes, 'in:': In,'Related names': Related_names, 'subjects': subjects,'places' : places}
        bdd = bdd.append(row, ignore_index = True)
```
