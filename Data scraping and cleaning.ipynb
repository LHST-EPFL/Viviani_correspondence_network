# Parsing Cited names

```python
#Clean cited names
bdd['Cited names'] = bdd['Cited names'].str.replace('\n', '', regex = False)
bdd['Cited names'] = bdd['Cited names'].str.replace('\xa0\xa0', '', regex = False)
bdd['Cited names'][4]
```

```python
#Remove destinatario
bdd['Cited names'] = bdd['Cited names'].str.replace('(.[^)]+)\(destinatario\.\)', ')', regex = True)
bdd['Cited names'] = bdd['Cited names'].str.replace('(^\))', '', regex = True)

#Remove commas
bdd['Cited names'] = bdd['Cited names'].str.replace(',', '', regex = False)

#Split by citato
bdd['Cited names'] = bdd['Cited names'].str.split('(citato.)',expand = False, regex = False)
bdd['Cited names'][4]
```

```python
#Removing last cited name because always ''
for i in range (0, len(bdd['Cited names'])):
    if isinstance(bdd['Cited names'][i], __builtins__.list) == True:
        bdd['Cited names'][i] = [x for x in bdd['Cited names'][i] if x != '']
```

```python
#format 1 : 1xxx-16xx / 1xxx-15xx / 1xxx-17xx
#format 2 : sec. ... (to split)
#format 3 : a.C.
#format 4 : n. 1590
#format 5 : m. 1669
def clean_name(name):
    l= name
    done = False
    while not done:
        if '-16' in l:
            l = l.rsplit(' ',1)[0]
        elif '-17' in l:
            l =  l.rsplit(' ',1)[0]
        elif '-15' in l:
            l =  l.rsplit(' ',1)[0]
        elif '-14' in l:
            l =  l.rsplit(' ',1)[0]
        elif '-13' in l:
            l =  l.rsplit(' ',1)[0]
        elif '-12' in l:
            l =  l.rsplit(' ',1)[0]
        elif '-11' in l:
            l =  l.rsplit(' ',1)[0]
        elif 'ca.' in l:
            l =  l.rsplit('ca.',-1)[0]
        elif 'a.C.' in l:
            l =  l.rsplit(' ', 2)[0]
        elif '-post' in l:
            l =  l.rsplit(' ', 2)[0]
        elif 'sec.' in l:
            l =  l.split('sec.', 1)[0]
        elif 'm.' in l:
            l =  l.split('m.', 1)[0]
        elif 'n.' in l:
            l =  l.split('n.', 1)[0]
        else:
            done = True
        
    return l
```

```python
#removes the birth/death year from the name to be the same as expeditor/destinator 
for index, row in bdd.iterrows():
    if isinstance(row['Cited names'], __builtins__.list) == True:
        for x in range(len(row['Cited names'])):
            bdd.at[index, 'Cited names'][x] = clean_name(bdd.loc[index]['Cited names'][x])
```

```python
#Extracting list of unique names
#appending all lists
list = []
for index, row in bdd.iterrows():
    if isinstance(row['Cited names'], __builtins__.list) == True:
        list.extend(bdd.loc[index]['Cited names'])
    
#turning the list to a set to have unique values
set_of_uniques = set(list)
print(set_of_uniques)

len(set_of_uniques)
```

```python
textfile = open("unique_names.txt", "w")
for element in set_of_uniques:
    textfile.write(element + "\n")
textfile.close()
```
